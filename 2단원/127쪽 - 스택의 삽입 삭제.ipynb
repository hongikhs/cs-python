```
class Stack(object):

    def __init__(self):
        self.items = []

    def __repr__(self):
        return repr(self.items)

    def push(self, x):
        self.items.append(x)

    def pop(self):
        return self.items.pop()

    def empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def top(self):
        return self.items[-1]
```

```
S = Stack()
```

```
S.push(1)
S.push(2)
S.push(3)
```
