```
class Queue(object):

    def __init__(self):
        self.items = []

    def __repr__(self):
        return repr(self.items)

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        return self.items.pop(0)

    def empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def front(self):
        return self.items[0]

    def back(self):
        return self.items[-1]
```

```
Q = Queue()
```
