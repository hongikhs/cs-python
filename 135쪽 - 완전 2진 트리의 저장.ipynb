```python
# 트리를 저장할 리스트 만들기()
# T = [None] * (1<<10)
T = [None] * 10
```

```python
T
```

```python
# 노드의 갯수
n = int(input())
```

```python
# 트리 입력받기
for i in range(1, n+1):
    T[i] = int(input())
```

```python
T
```

```python
# 트리 출력하기
k = 1
while k <= n:
    i = k
    while i <= n and i <= 2 * k - 1:
        print(T[i], end=' ')
        i += 1
    print()
    k *= 2
```
